### 1. IMPORTS

```python
from pathlib import Path
import sys
import pandas as pd
```

### 2. ORGANIZANDO ARQUIVOS 

```python
BASE_DIR = Path.cwd().parent
sys.path.append(str(BASE_DIR / "src" / "data_generation"))

from generate_custos import (
    calcular_custo_variavel, calcular_custo_fixo,
    gerar_custos_gerais, calcular_margem_bruta, salvar_arquivos
)

# Paths dos arquivos
DATA_RAW = BASE_DIR / "data/raw"
producao_path = DATA_RAW / "producao_mensal.xlsx"
custos_gerais_path = DATA_RAW / "custos_gerais.xlsx"
```

### 3. RODANDO SCRIPTS

```python
# 1. Carregar produção
df_prod = pd.read_excel(producao_path)
```

```python
# 2. Gerar custos operacionais por campo
df_prod = calcular_custo_variavel(df_prod, custo_var_brl_por_barril=50)
df_prod = calcular_custo_fixo(df_prod, custo_fixo_por_barris_dia=5000)
df_prod = calcular_margem_bruta(df_prod)
```

```python
# 3. Gerar custos gerais
df_custos_gerais = gerar_custos_gerais(start="2005-01", end="2025-12",
                                       custo_admin_base=50000,
                                       custo_manut_base=40000,
                                       custo_logistica_base=30000)
```

### 4. SALVAR ARQUIVOS

```python
salvar_arquivos(df_custos_gerais, df_prod, custos_gerais_path, producao_path)
```
