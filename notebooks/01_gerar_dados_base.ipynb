### 1. IMPORTS

```python
from pathlib import Path
import sys
import importlib.util
```

### 2. CONFIGURAR PASTAS

```python
BASE_DIR = Path.cwd().parent

def import_module_from_path(name, path):
    spec = importlib.util.spec_from_file_location(name, path)
    module = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(module)
    return module

preco_path = BASE_DIR / "generate_preco_petroleo.py"
cambio_path = BASE_DIR / "generate_cambio.py"
campos_path = BASE_DIR / "generate_campos.py"

#Criar pasta data/raw
DATA_RAW = BASE_DIR / "data/raw"
DATA_RAW.mkdir(parents=True, exist_ok=True)
```

### 3. IMPORTAR FUNÇÕES DOS SCRIPTS

```python
preco_module = import_module_from_path("generate_preco_petroleo", preco_path)
cambio_module = import_module_from_path("generate_cambio", cambio_path)
campos_module = import_module_from_path("generate_campos", campos_path)
```

### 4. Rodar Funções

```python
preco_module.gerar_preco_petroleo(output_path=DATA_RAW / "preco_petroleo.xlsx")
cambio_module.gerar_cambio(output_path=DATA_RAW / "cambio.xlsx")
campos_module.gerar_campos(output_path=DATA_RAW / "campos_petroliferos.xlsx", n_campos=10)
```
