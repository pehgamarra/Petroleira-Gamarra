# Notebook que faz a criação dos arquivos de produção mensal (.raw e .csv)

### 1. IMPORTS

```python
from pathlib import Path
import sys
import pandas as pd
```

### 2. ORGANIZANDO ARQUIVOS

```python
# Ajustar sys.path para importar script
BASE_DIR = Path.cwd().parent
sys.path.append(str(BASE_DIR / "src" / "data_generation"))
from generate_producao_mensal import (
    carregar_dados, merge_preco_cambio, gerar_producao_total,
    adicionar_custos_lucro, salvar_arquivos
)

# Paths dos arquivos
DATA_RAW = BASE_DIR / "data/raw"
campos_path = DATA_RAW / "campos_petroliferos.xlsx"
preco_path = DATA_RAW / "preco_petroleo.xlsx"
cambio_path = DATA_RAW / "cambio.xlsx"
csv_path = DATA_RAW / "producao_mensal_raw.csv"
excel_path = DATA_RAW / "producao_mensal.xlsx"
```

### 3. RODAR FUNÇÕES

```python
# Carregar dados
campos, preco, cambio = carregar_dados(campos_path, preco_path, cambio_path)
```

```python
# Juntar dados de preço e câmbio
preco_df = merge_preco_cambio(preco, cambio)
```

```python
# Gerar produção total
df = gerar_producao_total(campos, preco_df)
```

```python
# Adicionar custos e lucro
df = adicionar_custos_lucro(df)
```

```python
# Salvar arquivos
salvar_arquivos(df, csv_path, excel_path)
```
