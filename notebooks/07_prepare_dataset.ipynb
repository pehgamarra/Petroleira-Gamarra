### IMPORTS

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

### PREPARANDO ARQUIVOS

```python
BASE_DIR = Path.cwd().parent
sys.path.append(str(BASE_DIR / "src" / "ml"))

from prepare_dataset import prepare_ml_dataset
```

### RODAR PREPARAÇÃO


```python
df_ml = prepare_ml_dataset(BASE_DIR)
df_ml.head()
```

### VERIFICAÇÃO DE SPLITS

```python
print(df_ml["split"].value_counts())

sns.countplot(x="split", data=df_ml)
plt.title("Distribuição de amostras por split temporal")
plt.show()
```

### VERIFICAR TARGETS

```python
fig, ax = plt.subplots(2, 1, figsize=(12,6), sharex=True)
ax[0].plot(df_ml["data"], df_ml["target_producao_next"]/1e6)
ax[0].set_title("Target: Produção do Próximo Mês (10^6 barris)")
ax[1].plot(df_ml["data"], df_ml["target_receita_next"]/1e6, color="orange")
ax[1].set_title("Target: Receita do Próximo Mês (milhões BRL)")
plt.tight_layout()
plt.show()
```

### SALVAR VISUALIZAÇÃO RESUMIDA


```python
print("Colunas do dataset final:")
print(df_ml.columns.tolist())

print("\nResumo dos splits:")
print(df_ml.groupby("split")[["producao_total_barris","receita_total_brl","lucro_total_brl"]].mean().round(2))
```
